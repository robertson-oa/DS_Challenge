## Dataset Overview

This data set includes information about 60 Participants human behaviour recognition that recorded sensory data using devices such as smart phones (ipohone and andriods) and smart watches. Sensors included:acclerometer, gyroscope, magnetometer, watch acclerometer, locataion, audio etc.The labels in the dataset includes; physical activites(e.g Walking, running), Daily activities(eg: sleeping) Locations(eg: School, Work, Home). The dataset is in csv format which can be. <a href = 'https://drive.google.com/file/d/1-AyizDvWkVVsp_NjCYXp_h5PbbPVlQ7B/view?usp=sharing'>Downloaded here</a>.  
Visualization are created from preliminary wrangling and cleaning process to facilitate exploration analysis to discover usage pattern.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import glob

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
plt.rcParams['figure.figsize'] = 8, 6
base_color = sb.color_palette('colorblind')[0]
```

## INTRODUCTION
* Human Activity Recognition (HAR): Recognize the activity being performed by an individual at a given moment
* Human Context Recognition (HCR): Recognize the context (environment, location, interaction) thatthe activityis being performed.
* The combination of the two is important for the understanding of the human activities
* Wide variety of applications related to personal assistance and recommendations

```python
# load in the dataset into a pandas dataframe
path = '/home/bobby/Downloads/Data Science Challenge'
df = pd.concat(map(pd.read_csv, glob.glob(path + "/*.csv")))
```

## Smart watch Accelerometr: Walking


There seems to be a weak corrletaion between the gyroscope mean_x and gyroscope mean_y

```python
sb.regplot(x=df["proc_gyro:3d:mean_x"], y=df["proc_gyro:3d:mean_y"])
plt.xlabel("proc_gyro:3d:mean_x")
plt.ylabel("proc_gyro:3d:mean_y")
plt.title("Smart watch Accelerometr: Walking",fontsize=14, fontweight='semibold', y=1.015)
plt.show()
```

## A distribution plot of Phone Accelerometer and Gyroscope


From the plot the acceleration mean of Y as against gyroscope is normally distributed

```python
sb.countplot(data=df, x='proc_gyro:3d:mean_x', color=base_color);
plt.suptitle("Phone Accelerometer and Gyroscope", fontsize=14, fontweight='semibold', y=1.015)
plt.xlabel('RawAcc:3D:Mean Y', fontsize = 15);
plt.ylabel('Proc Gyro:MagnitudeStats:TimeEntropy', fontsize = 15);
```
